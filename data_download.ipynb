```python
code = 'HK.00700'
startdate='2022-09-11'
enddate='2022-09-18'
ktype = KLType.K_15M
ret, data, page_req_key = quote_ctx.request_history_kline(code, start=startdate, end=enddate, ktype = ktype, max_count=50) # 5 per page, request the first page
if ret == RET_OK:
    print(data)
else:
    print('error:', data)
while page_req_key != None: # Request all results after
    print('*************************************')
    ret, data, page_req_key = quote_ctx.request_history_kline(code, start=startdate, end=enddate, ktype = ktype, max_count=5, page_req_key=page_req_key) # Request the page after turning data
    if ret == RET_OK:
        print(data)
    else:
        print('error:', data)
print('All pages are finished!')
```
